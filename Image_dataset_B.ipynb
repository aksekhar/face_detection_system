```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import os,cv2,keras,glob
import pandas as pd
from PIL import Image
```

```python
current_dir = "/content/drive/MyDrive/dataset/"
training_images_path = os.path.join(current_dir, 'training_images')
training_images_path
```

```python
def detect_faces(image):
  gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

  # Load the cascade classifier for faces
  face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

  # Detect faces in the image
  faces = face_cascade.detectMultiScale(gray, 1.3, 5)

  # Return the detected faces
  return faces
```

```python
image_files = [f for f in os.listdir(training_images_path) if f.endswith('.jpg')]
```

```python
x_cord=[]
y_cord=[]
height=[]
width=[]
face_detected=[]
image_name=[]
```

```python
# Create an empty DataFrame
df = pd.DataFrame()

for image_file in image_files:
  image_path = os.path.join(training_images_path, image_file)
  #print(image_path)

  image = cv2.imread(image_path)

  #Detect faces in the image
  faces = detect_faces(image)

  j=0
  for x,y,w,h in faces:
    image=cv2.rectangle(image,(x,y),(x+w,y+h),(255, 0, 0),2)
    j=j+1
    resized=cv2.resize(image,(int(image.shape[1]/2),int(image.shape[0])))

    length=len(faces)
    if length==0:
        a=0
        b=0
        c=0
        d=0
        j=0
        name=0

    else:
        a=faces[0,0]
        b=faces[0,1]
        c=faces[0,2]
        d=faces[0,3]
        name=image_file


    x_cord.append(a)
    y_cord.append(b)
    width.append(c)
    height.append(d)
    face_detected.append(j)
    image_name.append(name)

```

```python
metadf = pd.DataFrame(x_cord, columns = ['x'])

metadf['y']=y_cord
metadf['w']=width
metadf['h']=height
metadf['Total_Faces']=face_detected
metadf['Image_Name']=image_name

metadf.head(100)
```
